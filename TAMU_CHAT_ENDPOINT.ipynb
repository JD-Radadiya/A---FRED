# Code to upload a file

```python
API_BASE = "https://chat-api.tamu.ai"

def upload_file_to_tamu(file_path: str, purpose: str = "fine-tune"):
    url = f"{API_BASE}/api/v1/files/"
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {API_KEY}",
    }

    files = {
        "file": open(file_path, "rb"),
        "purpose": (None, purpose),
    }

    response = requests.post(url, headers=headers, files=files)
    response.raise_for_status()
    data = response.json()
    print("Upload response:", json.dumps(data, indent=2))

    return data

upload_response = upload_file_to_tamu('/Users/jaydeepradadiya/Downloads/ITF Blog 2.pdf')
```

```python
API_BASE = "https://chat-api.tamu.ai"

def find_file_in_tamu(file_name: str):
    url = f"{API_BASE}/api/v1/files/search?filename={file_name}"
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {API_KEY}",
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    # print("File details:", json.dumps(data, indent=2))

    return data

find_file_in_tamu('ITF Blog 2.pdf')
```

```python
# code to access knowledge base from api

API_BASE = "https://chat-api.tamu.ai"

def knowledge_base_query_tamu():
    url = f"{API_BASE}/api/v1/knowledge/list"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {API_KEY}",
    } 

    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    # print("File details:", json.dumps(data, indent=2))

    return data
```

```python
def get_file_id_from_kb_name_and_filename(
    kb_list: list[dict],
    kb_name: str,
    file_name: str,
) -> str:
    """
    Given:
      - kb_list: result from knowledge_base_query_tamu()
      - kb_name: name of the knowledge base (e.g. 'API Test')
      - file_name: file's human-readable name (e.g. 'DL-14A.pdf')

    Returns:
      - file_id (str) if found

    Raises:
      - ValueError if kb or file cannot be found
    """
    # 1) Find the right knowledge base by name
    for kb in kb_list:
        if kb.get("name") == kb_name:
            # 2) Inside that KB, scan its 'files' list
            for f in kb.get("files", []):
                meta = f.get("meta", {})
                if meta.get("name") == file_name:
                    return f["id"]

            raise ValueError(
                f"File '{file_name}' not found in knowledge base '{kb_name}'. "
                f"Available files: {[f.get('meta', {}).get('name') for f in kb.get('files', [])]}"
            )

    raise ValueError(
        f"Knowledge base '{kb_name}' not found. "
        f"Available KBs: {[kb.get('name') for kb in kb_list]}"
    )

kb_list = knowledge_base_query_tamu()
file_id = get_file_id_from_kb_name_and_filename(
    kb_list,
    kb_name="API Test",
    file_name="DL-14A.pdf",
)
print("Found file ID:", file_id)
```

# Code to get a response

```python
API_BASE = "https://chat-api.tamu.ai/openai"

def chat_with_tamu(
    system_prompt: str,
    user_prompt: str,
    model: str = "protected.gpt-5",
    temperature: float = 0.2,
):
    url = f"{API_BASE}/chat/completions"
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json",
    }
    
    payload = {
        "model": model,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_prompt},
        ],
        "temperature": temperature,
        # try to hint no streaming; if they ignore it, we still handle SSE below
        "stream": False,
    }
    
    # stream=True so we can iterate line-by-line if it's SSE
    resp = requests.post(url, headers=headers, json=payload, stream=True)
    resp.raise_for_status()

    data = resp.json()
    return data["choices"][0]["message"]["content"]
```

```python
def chat_with_tamu_using_kb_file(
    kb_name: str,
    file_name: str,
    base_system_prompt: str,
    base_user_prompt: str,
    model: str = "protected.gpt-5",
    temperature: float = 0.2,
) -> str:
    """
    1) Looks up file_id for (kb_name, file_name)
    2) Augments system+user prompts with that file_id info
    3) Calls chat_with_tamu and returns the model's answer
    """
    file_id = get_file_id_from_kb_name_and_filename(kb_list, kb_name, file_name)

    # System prompt: tell the model it has access to this KB doc
    system_prompt = (
        base_system_prompt
        + "\n\nYou have access to a knowledge-base document with:\n"
        f"- knowledge_base_name: {kb_name}\n"
        f"- file_id: {file_id}\n"
        f"- filename: {file_name}\n"
        "Use the contents of that document when answering the user."
    )

    # User prompt: explicitly point to the same file
    user_prompt = (
        base_user_prompt
        + "\n\nThe document I want you to explain is the knowledge-base file with:\n"
        f"- knowledge_base_name: {kb_name}\n"
        f"- file_id: {file_id}\n"
        f"- filename: {file_name}\n"
        "Please base your explanation on that document."
    )

    return chat_with_tamu(
        system_prompt=system_prompt,
        user_prompt=user_prompt,
        model=model,
        temperature=temperature,
    )

system_prompt = "You are a helpful teaching assistant for CS courses at Texas A&M."
user_prompt = "Explain the PDF I have provided in simple terms."

kb_name = "API Test"        # whatever you named your KB
file_name = "DL-14A.pdf"    # the file in kb['files'][...]['meta']['name']

answer = chat_with_tamu_using_kb_file(
    kb_name=kb_name,
    file_name=file_name,
    base_system_prompt=system_prompt,
    base_user_prompt=user_prompt,
)

print(answer)
```
